```python
import pandas as pd 
import numpy as np
bank = pd.read_csv(r'/Users/sahilchavan/Downloads/Python(Gaurav sir)/all imports/bank_disbursals.csv')
bank.head()
```

```python
d = {"PL" : ["Unsec"] , "AL" : ["Sec"] , np.NaN : ["missing"]}
bank['using_map'] = bank.loc[:,"Product"].map(d)
bank.head(10)
```

```python
bank["using_where"] = np.where((bank["Product"]== "PL") & (bank["Loan"] > 100000),"HT",
                              np.where((bank["Product"] == "PL") & (bank["Loan"] < 50000) , "LT" ,
                                      np.where((bank["Product"] == "AL") & (bank["Loan"] > 80000),"HT",
                                              np.where((bank["Product"] == "AL") & (bank["Loan"] < 40000),"LT",
                                                      np.where((bank["Product"] == "TWL") & (bank["Loan"] > 70000),"HT",
                                                              np.where((bank["Product"] == "TWL") & (bank["Loan"]  < 30000),"LT",
                                                                      np.where(bank["Product"].isna(), "NA" ,"MT")))))))
```

```python
bank.head()
```

```python
def using_func(df):
    if df["Product"] == "PL":
        if df['Loan'] > 100000 : val = "HT"
        elif df["Loan"] < 50000 : val = "LT"
        else : val = "MT"
        
    elif df["Product"] == "AL":
        if df["Loan"] > 80000 : val = "HT"
        elif df["Loan"] < 40000 : val = "LT"
        else : val = "MT"
      
    elif df["Product"] == "TWL":
        if df["Loan"] > 70000 : val = "HT"
        elif df["Loan"] < 30000 : val = "LT"
        else : val = "MT"
    else : val = "NA"
        
    return val

        
```

```python
bank["using_function"] = bank.apply(using_func, axis = 1)
bank.head(10)
```

```python
import pandas as pd 
import numpy as np
bank = pd.read_csv(r'/Users/sahilchavan/Downloads/Python(Gaurav sir)/all imports/bank_disbursals.csv')
bank.head()
```
